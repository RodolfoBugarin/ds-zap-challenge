```python
import pandas as pd
import numpy as np
```

```python
import json 
from pandas.io.json import json_normalize
```

```python
import pickle
```

```python
#################################
# Converter JSON de Treino em DF
#################################
```

```python
d = []
with open('source-4-ds-train.json', encoding="utf8") as f:
    for line in f:
        d.append(json.loads(line))
```

```python
df = json_normalize(d)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace('.', '_')
```

```python
# Conferindo DF de treino
```

```python
pd.set_option('display.max_columns', 999)  ### para o pandas mostrar mais colunas.
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
df.groupby('address_city').address_city.count()
```

```python
# Gravar DF Treino em formato pickle

outfile = open('source-4-ds-train.pickle','wb')
pickle.dump(df, outfile)
outfile.close()
```

```python
# DF TO EXCEL

from pandas import ExcelWriter

writer = ExcelWriter('source-4-ds-train.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
